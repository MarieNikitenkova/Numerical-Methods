```
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```
def f(x, y):
    z = x ** 2 + y ** 2
    return np.sqrt((1 + 2 * (z ** 2) + np.sqrt(4 * (z ** 2) + 1))/(2 * (z ** 4)))
print(f(0, 0))
```

```
x = np.linspace(-1e-20, 1e-20, 100)
y = np.linspace(-1e-20, 1e-20, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z)

fig = plt.figure()
ax2 = fig.add_subplot(1, 2, 2, aspect='equal')
ax2.contourf(X, Y, Z)

plt.show()
```

Решаем неравенства f(z) >= 10 и f(z)>=100. Круги z=sqrt(11)/10 и z=sqrt(101)/100 есть решения этих неравенст соответственно.

```
x = lambda a: np.cos(a)   # e = 0.1
y = lambda a: np.sin(a)
fig = plt.subplots()
a = np.linspace(0, 2 * 3,141592 ,100)
plt.plot((np.sqrt(11)/10) * x(a), (np.sqrt(11)/10) * y(a))
plt.show()
```
