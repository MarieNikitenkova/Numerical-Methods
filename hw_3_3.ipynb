```
import numpy as np
import matplotlib.pyplot as plt

a = np.zeros ((15, 28))
a [2:-2, 1] = 1; a [2, 2:6] = 1
a [2:7, 6] = 1; a [7:-2, 7] = 1
a [7, 2:7] = 1; a [-3, 2:7] = 1
a [2:-2, 10] = 1; a [2:-2, 14] = 1
a [2:-2, 18] = 1; a [-3, 10:19] = 1
a [2:-2, 26] = 1; a [2, 21:26] = 1
a [7, 21:26] = 1; a [-3, 21:26] = 1
plt.imshow(a)
```

```
u, s, v = np.linalg.svd(a)
print(u)
print("*************")
print(s)
print("*************")
print(v)
```

```
np.linalg.matrix_rank(a)
for i in range(1, len(a)+1):
  aa = u[:,0:i] @ np.diag(s[0:i]) @ v[0:i,:]
  print("i =", i)
  plt.imshow(aa)
  plt.show()
```
