```python
import numpy as np

from scipy.optimize import minimize
```

```python
with np.load('/data_89.npz') as data:
    M, y = data['A2'], data['y2']
```

```python
def f(x, a):
    return 1 / (np.exp(a[0] + x@a[1:]) + 1)
```

```python
def optimize(a):
    return np.sum((f(M, a) - y)**2)
```

```python
n = M.shape[1]
a_array = minimize(optimize, x0=np.zeros(n+1))
print(a_array)
```
