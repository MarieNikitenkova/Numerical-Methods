```python
import numpy as np
import matplotlib.pyplot as plt


def plot(A):
  U, S, V = np.linalg.svd(A)
  v = np.linspace(0, 2 * np.pi, 100)
  x = np.cos(v)
  y = np.sin(v)
  x1 = (A[0][0] * x + A[0][1] * y)
  y1 = (A[1][0] * x + A[1][1] * y)
  
  v1 = V.T[0]
  v2 = V.T[1]
  u1 = U[:,0][:2]
  u2 = U[:,1][:2]
  
  origin = np.array([[0, 0],[0, 0]])
  
  fig= plt.figure(figsize=(5,5))
  p1 = plt.gca()
  plt.title('Right')
  p1.plot(x, y)
  p1.quiver(*origin,[v1[0], v2[0]],[v1[1],v2[1]],scale_units='xy',scale=1)
  
  fig= plt.figure(figsize=(5,5))
  p2 = plt.gca()
  plt.title('Left')
  p2.plot(x1, y1)
  k = np.array([S[0]*u1, S[1]*u2])
  print(k)
  
  p2.quiver(*origin, k[:,0], k[:,1] ,scale_units='xy', scale = 1)

  plt.show()
```

```python
A = np.array([[3,0],[0,(-2)]])
plot(A)
B = np.array([[1,1],[1,1]])
plot(B)
```
