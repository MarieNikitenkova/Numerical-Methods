```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def exp_sample(n):
    y = np.random.uniform(0, 1, n)
    x = -np.log(1 - y)
    return x
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 7))
xx = np.linspace(0, 10, 1000)
yy = np.exp(-xx)
for i, n in enumerate([10, 100, 500, 1000, 2000, 5000]):
    x = exp_sample(n)
    axs[i // 3, i % 3].hist(x, bins=int(1 + 3.322*np.log(n)), density=True)
    axs[i // 3, i % 3].plot(xx, yy)
    axs[i // 3, i % 3].set_title(f'n={n}')
```

С увеличением размера выборки гистограмма больше соответсвует ожидаемой плотности вероятности

```python
from scipy.stats import kstest

x = exp_sample(10**8)
pvalue = kstest(x, lambda x: 1 - np.exp(-x))[1]
print(pvalue)
```
