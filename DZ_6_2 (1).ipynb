```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw(n):
    x = np.linspace(-1, 1, n)
    y = np.cos(n*x)
    P = np.polyfit(x, y, n)
    plt.plot(x,y,'o')
    plt.plot(x, np.polyval(P,x))
    plt.grid(True)
    plt.show()
draw(16)
draw(32)
draw(64)
```
